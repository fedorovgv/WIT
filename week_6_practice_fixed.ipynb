Pandas это модуль, или по-другому библиотека, которая содержит в себе различные функции и классы, которые мы можем вызывать и использовать. Слово `import` говорит, что мы хотим загрузить библиотеку для использования, `as` говорит, что мы хотим сократить название библиотеки до двух букв `pd`.
