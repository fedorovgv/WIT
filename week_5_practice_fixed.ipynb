Пример работы с series, получим именования пассажиров. 

```python
name = data['Name'].iloc[0]
name
```

```python
splited_name = name.split(' ')
splited_name
```

```python
address = splited_name[1]
address
```

```python
address = address[:-1]
address
```

```python
def get_address(name):
  splited_name = name.split(' ')
  address = splited_name[1]
  address = address[:-1]
  return address
```

```python
data['Address'] = data['Name'].apply(get_address)
data
```
